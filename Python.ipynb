```python
import math

class mathObject():
    
    def __init__(self, int_a, int_b, int_c):
        self.int1 = int_a
        self.int2 = int_b
        self.int3 = int_c
        
    def avg(self):
        return float(self.sum()) / 3
    
    def sum(self):
        return self.int1 + self.int2 + self.int3
    
    def least(self):
        return min(self.int1, self.int2, self.int3)
    
    def mult_sum_by_arg(self, multiplier):
        return self.sum() * multiplier
    
    def __len__(self):
        return self.sum()
    
obj1 = mathObject(3, 5, 7)
print(obj1.sum())
print(obj1.avg())
print(obj1.least())
print(obj1.mult_sum_by_arg(.3))


obj12 = mathObject(9, -3, 5)
print(obj12.sum())
print(obj12.avg())
print(obj12.least())
print(obj12.mult_sum_by_arg(.3))

print(len(obj1))
```
